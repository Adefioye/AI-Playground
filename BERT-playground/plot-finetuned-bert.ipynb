```python
import matplotlib.pyplot as plt
import numpy as np

# Model names grouped by base model (stripped for X-axis)
base_models = [
    "deberta-v3-large",
    "distilbert-base-uncased",
    "ModernBERT-large",
    "ModernBERT-Large-Instruct"
]

# Accuracy scores
original_accuracies = [0.4700, 0.5619, 0.4700, 0.4549]
finetuned_accuracies = [0.5300, 0.9323, 0.9495, 0.9158]

# Bar chart setup
x = np.arange(len(base_models))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2, original_accuracies, width, label='Original')
bars2 = ax.bar(x + width/2, finetuned_accuracies, width, label='Finetuned (sarcasm)')

# Labels and titles
ax.set_ylabel('Accuracy')
ax.set_title('Model Accuracy Before and After Finetuning on Sarcasm')
ax.set_xticks(x)
ax.set_xticklabels(base_models, rotation=45, ha="right")
ax.legend()

# Add accuracy values on top of bars
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # Offset text
                    textcoords="offset points",
                    ha='center', va='bottom')

add_labels(bars1)
add_labels(bars2)

plt.tight_layout()
plt.show()
```

```python
from scipy import stats
import numpy as np

# Example accuracy numbers for two different evaluations (e.g., Model A vs. Model B)
# These represent the accuracy on the same set of instances for each evaluation.
# Assume these are lists of accuracy scores from multiple runs or folds.
accuracy_eval1 = np.array([0.85])
accuracy_eval2 = np.array([0.84])

# Perform the paired t-test
t_statistic, p_value = stats.ttest_rel(accuracy_eval1, accuracy_eval2)

print(f"Paired t-test results:")
print(f"  t-statistic: {t_statistic:.4f}")
print(f"  p-value: {p_value:.4f}")

# Interpret the results (common significance level alpha = 0.05)
alpha = 0.05
if p_value < alpha:
    print("\nConclusion: The difference in accuracy between the two evaluations is statistically significant.")
else:
    print("\nConclusion: The difference in accuracy between the two evaluations is not statistically significant.")
```
